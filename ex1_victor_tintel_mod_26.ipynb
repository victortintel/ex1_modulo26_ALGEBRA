```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
```

```python
df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect('equal', 'box')
ax.set(xlim=(0, 11), ylim=(0, 11))

np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0
p2

df = pd.DataFrame({'p1': p1, 'p2': p2})

plt.scatter(x = df.p1, y=df.p2);
```

```python
df['média'] = df.mean(axis=1)
df['evolução'] = df.p2-df.p1
df.style.format(precision=1, decimal=',')
```

Quando temos os dados das provas de alunos, no caso são 49 alunos, podemos tirar uma conclusão com relação a evolução. Podemos ver que em grande maioria não teve diferença de 2 pontos para mais ou para menos de cada aluno, portanto uma escola  nas próximas provas dar uma ajuda extra para alunos que tiraram nota baixa sabendo que em grande maioria se deixar como está, só aumentará no máximo 2 pontos para cima.
